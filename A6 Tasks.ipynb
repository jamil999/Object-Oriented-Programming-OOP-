```python
#task_1
class Student:
    ID=0
    def __init__(self,name,department,age,cgpa):
        self.name=name
        self.department=department
        self.age=age
        self.cgpa=cgpa
        Student.ID=Student.ID+1
        
    def get_details(self):
        print('ID:',self.ID)
        print('Name:',self.name)
        print('Department:',self.department)
        print('Age:',self.age)
        print('CGPA:',self.cgpa)
        
    @classmethod
    def from_String(cls,parameters):
        name,department,age,cgpa=parameters.split('-')
        return Student(name,department,age,cgpa)

s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()

print('\nA class variable is defined in the class construction and assigns value in the class declaration. It belongs to the class and not to the object.On the other hand, an instance variable assigns value inside class methods. It is only owned by instances of the class. These are the the differences between a class variable and an instance variable.\n')
print('An instance method can access the attributes and the other methods through the "self" parameter. It can be only invoked through an instance of the class. But a class method takes a different parameter "cls" instead of "self" that points to the class instead of the object. These are the differences between an instance method and class method.')
```

```python
#task_2
class Dates:
    def __init__(self,time):
        self.time=time
        
    @classmethod
    def toDashDate(cls,parameter):
        new_time=parameter.replace('/','-')
        return Dates(new_time)
    
    def getDate(self):
        return self.time

date1 = Dates("05-09-2020")
dateFromDB = "05/09/2020"
date2= Dates.toDashDate(dateFromDB)
if(date1.getDate() == date2.getDate()):
 print("Equal")
else:
 print("Unequal")

print('\nIn the conditional statement it prints "Equal" because in date1 object the value 05-09-2020 was passed in the class. In date2 the "05/09/2020" was passed to the class mathod "toDashDate" which replaced the / into - and made the value into "05-09-2020". Now the getDate method returns date1 and date2 objects value and as now both are the same which is 05-09-2020, the conditional statement prints Equal.')
```

```python
#task_5
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        
    @classmethod
    def personByBirth(cls,name,year):
        age_1=2020-year
        return Person(name,age_1)
        
    @staticmethod
    def adultCheck(age,gender):
        if 18>age:
            if gender=='male':
                return 'He is not adult'
            if gender=='female':
                return 'She is not adult'
        
        if 18<=age:
            if gender=='male':
                return 'He is adult'
            if gender=='female':
                return 'She is adult'       

person1 = Person('Alice', 21)
person2 = Person.personByBirth('Bob', 1996)
print(person1.age)
print(person2.age)
print(person1.name)
print(person2.name)
print(Person.adultCheck(22, "male"))
print(Person.adultCheck(12, "female"))
```

```python
#task_6
class Fruit:
    fruitCount=0
    def __init__(self,name,count):
        self.name=name
        self.count=count
        Fruit.fruitCount=Fruit.fruitCount+count
        
    @classmethod
    def saySomethingGood(cls):
        print('Fruits are good for health')
        
    @classmethod
    def resetcount(cls):
        Fruit.fruitCount=0

apples = Fruit("Apple", 3);
pears = Fruit("Pear", 4);
print(apples.name, apples.count)
print(pears.name,pears.count)
print("Total number of fruits", Fruit.fruitCount)
Fruit.saySomethingGood()
Fruit.resetcount()
print("Total number of fruits", Fruit.fruitCount)
```

```python
#task_7
class Cat:
    Number_of_cats=0
    def __init__(self,colour,doing):
        self.colour=colour
        self.doing=doing
        Cat.Number_of_cats=Cat.Number_of_cats+1
        
    @classmethod
    def no_parameter(cls):
        return Cat('White','sitting')
    
    @classmethod
    def first_parameter(cls,new_colour):
        return Cat(new_colour,'sitting')
    
    @classmethod
    def second_parameter(cls,new_doing):
        return Cat('Grey',new_doing)
    
    def changeColor(self,new):
        self.colour=new
        
    def printCat(self):
        print(self.colour,'cat is',self.doing)
        
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#task_8
import math
class Cylinder:
    radius=5
    height=18
    def __init__(self,up_radius,up_height):
        print('Default radius='+str(self.radius)+' and height='+str(self.height))
        Cylinder.radius=up_radius
        Cylinder.height=up_height
        print('Updated: radius='+str(self.radius)+' and height='+str(self.height))
    
    @classmethod
    def swap(cls,height,radius):
        return Cylinder(radius,height)
    
    @classmethod
    def changeFormat(cls,string):
        radius,height=string.split('-')
        return Cylinder(float(radius),float(height))
    
    @staticmethod
    def area(val1,val2):
        result=(2*math.pi*val1**2)+(2*math.pi*val1*val2)
        print('Area:',result)
    
    @staticmethod
    def volume(val1,val2):
        result=math.pi*val1**2*val2
        print('Volume:',result)
    
c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```
