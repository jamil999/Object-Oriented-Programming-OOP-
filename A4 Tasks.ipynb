```python
#task_1
class Calculator:
    def __init__(self,value_1,operator,value_2):
        self.first="Let's Calculate!" 
        self.value_1=value_1
        self.operator=operator
        self.value_2=value_2
    def add(self,value_1,operator,value_2):
        result=value_1+value_2
        print('result:',result)
    def subtract(self,value_1,operator,value_2):
        result=value_1-value_2
        print('result:',result)
    def multiply(self,value_1,operator,value_2):
        result=value_1*value_2
        print('result:',result)
    def divide(self,value_1,operator,value_2):
        result=value_1/value_2
        print('result:',result)
        

value_1=int(input('Enter the first value:'))
operator=input('Enter the operator:')
value_2=int(input('Enter the second value:'))
object1=Calculator(value_1,operator,value_2)
print(object1.first)
print('Value 1:',object1.value_1)
print('Operator:',object1.operator)
print('Value 2:',object1.value_2)

if operator=='+':
    object1.add(object1.value_1,object1.operator,object1.value_2)
if operator=='-':
    object1.substract(object1.value_1,object1.operator,object1.value_2)
if operator=='*':
     object1.multiply(object1.value_1,object1.operator,object1.value_2)
if operator=='/':
    object1.divide(object1.value_1,object1.operator,object1.value_2)      
```

```python
#task_2
class Customer:
    def __init__(self,name):
        self.name=name
    def greet(self,name=''):
        if name=='':
            print('Hello!')
        else:
            print('Hello '+name+'!')
    def purchase(self,*product):
        total=len(product)
        print(self.name+', you purchased '+str(total)+' item(s)')
        for index in product:
            print(index)
            
customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
#task_3
class Panda:
    def __init__(self,name,gender,age):
        self.name=name
        self.gender=gender
        self.age=age
    def sleep(self,sleep_hours=0):
        if 3<=sleep_hours<=5:
            return '{} sleeps {} hours daily and should have Mixed Veggies'.format(self.name,sleep_hours)
        if 6<=sleep_hours<=8:
            return '{} sleeps {} hours daily and should have Eggplant & Tofu'.format(self.name,sleep_hours)
        if 9<=sleep_hours<=11:
            return '{} sleeps {} hours daily and should have Broccoli Chickhen'.format(self.name,sleep_hours)
        else:
            return '{} duration is unknown thus should have only bamboo leaves'.format(self.name+"'s")
        
panda1 = Panda("Kunfu","Male", 5)
panda2=Panda("Pan Pan","Female",3)
panda3=Panda("Ming Ming","Female",8)
print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())
```

```python
#task_5
class Vehicle:
    def __init__(self,x_coord=0,y_coord=0):
        self.x_coord=x_coord
        self.y_coord=y_coord
    def moveUp(self):
        self.y_coord=self.y_coord+1
    def moveLeft(self):
        self.x_coord=self.x_coord-1
    def moveRight(self):
        self.x_coord=self.x_coord+1
    def moveDown(self):
        self.y_coord=self.y_coord-1
    def print_position(self):
        current_cord=[]
        current_cord.append(self.x_coord)
        current_cord.append(self.y_coord)
        print(tuple(current_cord))
        
car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```

```python
#task_6
class Programmer:
    def __init__(self,programmer,know_language,total_experience):
        self.first='Horray! A new programmer is born'
        self.programmer=programmer
        self.know_language=know_language
        self.total_experience=total_experience
    def addExp(self,number):
        self.first='Updating experience of '+self.programmer
        self.total_experience=self.total_experience+number
    def printDetails(self):
        print(self.first)
        print('Name:',self.programmer)
        print('Language:',self.know_language)
        print('Experience:',self.total_experience,'years.')

p1 = Programmer("Ethen Hunt", "Java", 10)
p1.printDetails()
print('--------------------------')
p2 = Programmer("James Bond", "C++", 7)
p2.printDetails()
print('--------------------------')
p3 = Programmer("Jon Snow", "Python", 4)
p3.printDetails()
p3.addExp(5)
p3.printDetails()
```

```python
#task_7
class Student:
    def __init__(self,Name,ID,Department='CSE'):
        self.name=Name
        self.ID=ID
        self.department=Department
    def dailyEffort(self,Daily_effort):
        self.daily_effort=Daily_effort
        if 2>=Daily_effort:
            self.suggest='Should give more effort!'
        if 2<Daily_effort<=4:
            self.suggest='Keep up the good work!'
        if 4<Daily_effort:
            self.suggest='Excellent! Now motivate others.'
            
    def printDetails(self):
        print('Name:',self.name)
        print('ID:',self.ID)
        print('Department:',self.department)
        print('Daily Effort:',self.daily_effort,'hour(s)')
        print('Suggestion:',self.suggest)

harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```python
#task_8
class Patient:
    def __init__(self,patient,age,symptoms=''):
        self.patient=patient
        self.age=age
        self.symptoms=symptoms
    def add_Symptom(self,*symptoms):
        for index in symptoms:
            if index==symptoms[-1]:    
                self.symptoms=self.symptoms+index
            else:
                self.symptoms=self.symptoms+index+', '
    def printPatientDetail(self):
        print('Name:',self.patient)
        print('Age:',self.age)
        print('Symptoms:',self.symptoms)
p1 = Patient("Thomas", 23)
p1.add_Symptom("Headache")
p2 = Patient("Carol", 20)
p2.add_Symptom("Vomiting", "Coughing")
p3 = Patient("Mike", 25)
p3.add_Symptom("Fever", "Headache", "Coughing")
print("=========================")
p1.printPatientDetail()
print("=========================")
p2.printPatientDetail()
print("=========================")
p3.printPatientDetail()
print("=========================")
```

```python
#task_9
class Avengers:
    def __init__(self,name,partner,super_power=''):
        self.name=name
        self.partner=partner
        self.super_power=super_power
    def super_powers(self,*super_power):
        for index in super_power:
            if index==super_power[-1]:    
                self.super_power=self.super_power+index
            else:
                self.super_power=self.super_power+index+', '
    def printAvengersDetail(self):
        print('Name:',self.name)
        print('Partner:',self.partner)
        print('Super powers:',self.super_power)

a1 = Avengers('Captain America', 'Bucky Barnes')
a1.super_powers('Stamina', 'Slowed ageing')
a2 = Avengers('Doctor Strange', 'Ancient One')
a2.super_powers('Mastery of magic')
a3 = Avengers('Iron Man', 'War Machine')
a3.super_powers('Genius level intellect', 'Scientist ')
print("=========================")
a1.printAvengersDetail()
print("=========================")
a2.printAvengersDetail()
print("=========================")
a3.printAvengersDetail()
print("=========================")
```

```python
#task_10
class Shinobi:
    def __init__(self,Name,Rank,Number_of_mission=0,Salary=0):
        self.Name=Name
        self.Rank=Rank
        self.Number_of_mission=Number_of_mission
        self.Salary=Salary
    def calSalary(self,number):
        self.Number_of_mission=number
        if self.Rank=='Genin':
            self.Salary=(number*50)
        if self.Rank=='Chunin':
            self.Salary=(number*100)
        else:
            self.Salary=(number*500)
    def changeRank(self,change_rank):
        self.Rank=change_rank
    def printInfo(self):
        print('Name:',self.Name)
        print('Rank:',self.Rank)
        print('Number of mission:',self.Number_of_mission)
        print('Salary:',self.Salary)

naruto = Shinobi("Naruto", "Genin")
naruto.calSalary(5)
naruto.printInfo()
print('====================')
shikamaru = Shinobi('Shikamaru', "Genin")
shikamaru.printInfo()
shikamaru.changeRank("Chunin")
shikamaru.calSalary(10)
shikamaru.printInfo()
print('====================')
neiji = Shinobi("Neiji", "Jonin")
neiji.calSalary(5)
neiji.printInfo()
```
